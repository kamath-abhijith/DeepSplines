```python
import sys
sys.path.append('./../')

import os
import torch

import numpy as np

from torch import nn as nn
from torch.nn import functional as F
from matplotlib import pyplot as plt

from utils import *

plt.style.use(['science','ieee'])

plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["cm"],
    "mathtext.fontset": "cm",
    "font.size": 24})
```

```python
l = 5
x = torch.arange(-l,l,0.01)

relu = nn.ReLU()
leakyrelu = nn.LeakyReLU(0.1)
sigmoid = nn.Sigmoid()
soft = nn.Softshrink(lambd=1)
```

```python
plt.figure(figsize=(6,6))
ax = plt.gca()
# plot_hline(0, ax)
plot_signal(x, relu(x), ax=ax,
    plot_colour='blue', line_width=2,
    legend_label=r'$\max\{x,0\}$', legend_loc='lower right',
    xaxis_label=r'$x$',
    legend_show=False, n_col=1,
    xlimits=[-l+1,l-1], ylimits=[-l,l],
    save='relu')

plt.figure(figsize=(6,6))
ax = plt.gca()
plot_hline(0, ax)
plot_signal(x, leakyrelu(x), ax=ax,
    plot_colour='blue', line_width=2, line_style='-',
    xaxis_label=r'$x$',
    legend_label=r'$\max\{x,0.1x\}$', legend_loc='lower right',
    legend_show=False, n_col=1,
    xlimits=[-l+1,l-1], ylimits=[-l,l],
    save='leakyrelu')

plt.figure(figsize=(6,6))
ax = plt.gca()
plot_hline(0, ax)
plot_signal(x, soft(x), ax=ax,
    plot_colour='blue', line_width=2,
    xaxis_label=r'$x$',
    legend_label=r'$\text{soft}(x,1)$', legend_loc='lower right',
    legend_show=False, n_col=1,
    xlimits=[-l+1,l-1], ylimits=[-l,l],
    save='soft')

plt.figure(figsize=(6,6))
ax = plt.gca()
plot_hline(0, ax)
plot_signal(x, sigmoid(x), ax=ax,
    plot_colour='blue', line_width=2,
    xaxis_label=r'$x$',
    legend_label=r'$\sigma(x)$', legend_loc='lower right',
    legend_show=False, n_col=1,
    xlimits=[-l+1,l-1], ylimits=[-l,l])
```
